```python
#filter
import scipy
from scipy.signal import butter, lfilter
from scipy import signal
def filters (chanel,pole,low,high):
  
    b_notch, a_notch = signal.iirnotch(notch_freq, quality_factor, samp_freq)
    dn = signal.filtfilt(b_notch, a_notch, chanel)
    b, a = scipy.signal.butter(pole, [low, high], 'band', fs=200)
    df = scipy.signal.lfilter(b, a, dn)
    dframe = pd.DataFrame(df)
    return dframe


def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut 
    high = highcut 
    b, a = butter(order, [low, high], btype='band', fs=200)
    return b, a


def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y
```

## WINDOWING

```python
n_steps = 5
n_window = 40
n_chanel = 8
def split_sequences(sequences, n_steps, n_window):
    X, Y = list(), list()
    for i in range(len(sequences)):
        end_ix = i*n_steps + n_window
        if end_ix > len(sequences-n_window):
            break
        seq_x, seq_y = sequences[i*n_steps:end_ix, :-1], sequences[i*n_steps:end_ix,-1]
#         rms = np.sqrt(np.mean(np.square(seq_x),axis=0))
        out=np.mean(seq_y)
        X.append(seq_x)
        Y.append(out)
    X=np.array(X)
    Y=np.array(Y)
    
    return X, Y
```

## **TUNING DENGAN OPTUNA**

```python
import optuna

cvsplit = 5

def objective(trial):
  r2cv, rmsecv = list(), list()
  kf = KFold(n_splits=cvsplit)
  for train, test in kf.split(rms):
      rms_train, rms_test, y_train1, y_test1     = rms[train,:], rms[test,:], y1[train], y1[test] 
      rms_train, rms_test, y_train2, y_test2     = rms[train,:], rms[test,:], y2[train], y2[test] 
      rms_train, rms_test, y_train3, y_test3     = rms[train,:], rms[test,:], y3[train], y3[test] 
      rms_train, rms_test, y_train4, y_test4     = rms[train,:], rms[test,:], y4[train], y4[test] 
      rms_train, rms_test, y_train5, y_test5     = rms[train,:], rms[test,:], y5[train], y5[test] 

      rms_train = rms_train
      y_train = np.hstack((y_train1, y_train2, y_train3, y_train4, y_train5))
      rms_test = rms_test
      y_test = np.hstack((y_test1, y_test2, y_test3, y_test4, y_test5))

      if reg == "Random Forest":
            # RANDOM FOREST
            #List Hyperparameters that we want to tune.
            n_estimators = trial.suggest_int('n_estimators', 100, 300)
            max_depth = int(trial.suggest_float('max_depth', 1, 50, log=True))
            
            regr = RandomForestRegressor(n_estimators = n_estimators, max_depth=max_depth, random_state=0, bootstrap=True,)
            regr.fit(rms_train, y_train)
            y_pred = regr.predict(rms_test)

  hasil_r2 = r2_score(y_pred,y_test)
  hasil_mse = mean_squared_error(y_pred,y_test)
  hasil_rmse = np.sqrt(mean_squared_error(y_pred,y_test))
  r2cv.append(hasil_r2), rmsecv.append(hasil_rmse)
        #r2cv, rmsecv = np.array(r2cv), np.array(rmsecv)
  return np.mean(np.array(r2cv))
```

```python
# Tuning Parameter Random Forest
model = ["Random Forest","Decision Tree","kNN","SVM"]
reg = model[0] # indeks 0 untuk random forest

n_trials = 50
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials = n_trials)

trial_RF = study.best_trial
print(f"Hasil tuning {reg}")
print('R2: {}'.format(trial_RF.value))
print("Best hyperparameters: {}".format(trial_RF.params))

optuna.visualization.plot_optimization_history(study)
optuna.visualization.plot_slice(study)
```
